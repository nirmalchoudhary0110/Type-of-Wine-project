```python
total_nulls = df.isnull().sum().sum()

percent_nulls = (total_nulls / 37079) * 100
print(f"Percentage of null values: {percent_nulls:.2f}%")
```

```python
total_rows = df.shape[0]

total_duplicates = df.duplicated().sum()

percent_duplicates = (total_duplicates / total_rows) * 100
print(f"Percentage of duplicate rows: {percent_duplicates:.2f}%")
```

```python
col = 'alcohol'  

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df[col] < lower) | (df[col] > upper)]

outlier_count = outliers.shape[0]

total_points = df.shape[0]

percent_outliers = (outlier_count / total_points) * 100

print(f"Percentage of outliers in '{col}': {percent_outliers:.2f}%")
```

```python
x=df.iloc[:,1:7]
y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_test
```

```python
x_train
```

```python
y_train
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
Rr=LogisticRegression()
Rr.fit(x_train,y_train)
y_pred=Rr.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
a=Rr.predict(scaler.transform([[0.08,1.9,0.99651,3.42,0.82,9.5]]))
a
```

# Decision Tree Categorical(Post Pruning)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
C1=DecisionTreeClassifier()
C1.fit(x_train,y_train)
y_pred=C1.predict(x_test)
score1=accuracy_score(y_test,y_pred)
print("Accuracy =",score1)
```

```python
b=C1.predict(scaler.transform([[0.08,1.9,0.99651,3.42,0.82,9.5]]))
b
```

# Decision Tree Categorical(Pre Pruning)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
C2= DecisionTreeClassifier()
parameter={'criterion':['gini','entropy','log_loss'],
           'splitter':['best','random'],
           'max_features':['sqrt','log2'],
           'ccp_alpha':[0.0, 0.001, 0.01, 0.02, 0.05]}
C2= GridSearchCV(C2, parameter, cv=5)
C2.fit(x_train, y_train)

print(C2.best_params_)

y_pred = C2.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
c=C2.predict(scaler.transform([[0.08,1.9,0.99651,3.42,0.82,9.5]]))
c
```

# Random Forest Categorical

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
C3=RandomForestClassifier()
C3.fit(x_train, y_train)
y_pred=C3.predict(x_test)
score1=accuracy_score(y_test,y_pred)
print("Accuracy =",score1)
```

```python
d=C3.predict(scaler.transform([[0.08,1.9,0.99651,3.42,0.82,9.5]]))
d
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB 
C4= GaussianNB()

C4.fit(x_train,y_train) 
from sklearn.metrics import accuracy_score 
y_pred=C4.predict(x_test)
score =accuracy_score(y_pred,y_test)
print("%.2f"% score)
```

# SVM Algorithm

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.svm import SVC
C5= SVC(kernel='linear') 

C5.fit(x_train, y_train)
y_pred =C5.predict(x_test)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
